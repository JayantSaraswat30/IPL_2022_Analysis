### Loading the Libraries and Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('IPL.csv')
df.head()
```

### Basic Information

```python
df.info()
```

```python
print(f"The number of rows are {df.shape[0]} and the number of columns are {df.shape[1]}")
```

```python
df.isnull().sum()
```

```python
#1.Which team won the most matches?
match_wins = df['match_winner'].value_counts()
match_wins
```

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index,x = match_wins.values,palette='viridis')
plt.title("Most match win by team ")
```

```python
#Toss Decision
sns.countplot(x = df['toss_decision'], palette='viridis')
plt.title('Toss Decision Trends')
```

```python
#Toss vs Win
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count * 100)/df.shape[0]
percentage.round(2)
```

```python
#How do teams win? (Runs vs Wickets)
sns.countplot(x = df['won_by'])
plt.title('Won By')
```

```python
#Most "Player of the Match" Awards
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x = count.values, y = count.index, palette='mako')
plt.title('Player of the match')
```

```python
# Top Scorers
df.head()
```

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(5)
high.head()
```

```python
high.plot(kind= 'barh')
plt.title('Top 5 Scorer')
```
